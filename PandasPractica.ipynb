```python
import pandas as pd
```

```python
df = pd.read_csv("C:/Users/B725218/Downloads/Datos_Ventas_Tienda.csv")
df
```

```python
df2 = pd.read_csv("C:/Users/B725218/Downloads/Datos_Ventas_Tienda2.csv")
df2
```

```python
df_concatenado = pd.concat([df,df2])
df_concatenado
```

```python
df_concatenado.info()
```

```python
type(df_concatenado['Fecha'][0])
```

```python
df_concatenado['Fecha'] = pd.to_datetime(df_concatenado['Fecha'], format = '%d-%m/-%y')
df_concatenado
```

```python
df_final = df_concatenado.dropna()
df_final = df_final.drop_duplicates(subset=["Fecha","Producto","Cantidad","Precio Unitario","Total Venta"])
df_final
```

```python
df_agrupado = df_final.groupby("Producto")["Cantidad"].sum().sort_values(ascending=False).head(1)
df_agrupado
```

```python
meses = []
for i in df_final["Fecha"]:
    meses.append(i.month)
df_final["Meses"] = meses
mes_con_mas_ventas = df_final.groupby("Meses")["Total Venta"].sum().sort_values(ascending=False).head(1)
mes_con_mas_ventas
```

```python
venta_por_categoria = df_final.groupby("Producto")["Total Venta"].sum().sort_values(ascending=False)
venta_por_categoria
```

```python
df_final.to_csv("C:/Users/B725218/Desktop/Df_Final.csv")
```
